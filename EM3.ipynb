```python
# write code here

#variable allocation
months = {
    "1":"January","2":"February","3":"March",
    "4":"April","5":"May","6":"June",
    "7":"July","8":"August","9":"September",
    "10":"October","11":"November","12":"December",
    "Jan":"January","Feb":"February","Mar":"March",
    "Apr":"April","Jun":"June",
    "Jul":"July","Aug":"August","Sep":"September",
    "Oct":"October","Nov":"November","Dec":"December",
    "January":"January","February":"February","March":"March",
    "April":"April","May":"May","June":"June",
    "July":"July","August":"August","September":"September",
    "October":"October","November":"November","December":"December"
}

#function for cleaniing month data
def clean_month(month):
    for m in months:
        if month==m:
            return str(months[m])
            continue
        else:
            continue

#vectorize
recweek["birthmonth"] = recweek["birthmonth"].apply(clean_month)

#dump
recweek
```

```python
#write code here
#check type
recweek['birthyear'].dtype

#change int64 to str
recweek["birthyear"] = recweek["birthyear"].astype(str)

#variable allocation
years = r"^\d{2}$"
courses = {
    "BS LM":"BS Legal Management","Legal Management":"BS Legal Management","LM":"BS Legal Management",
    "BS MGT":"BS Management","Management":"BS Management","MGT":"BS Management",
    "BS CTM":"BS Communications Technology","Communications Technology":"BS Communications Technology","CTM":"BS Communications Technology",
    "BS ITE":"BS IT Entrepreneurship","Information Technology Entrepreneurship":"BS IT Entrepreneurship","ITE":"BS IT Entrepreneurship",
    "BS ME":"BS Management Engineering","Management Engineering":"BS Management Engineering","ME":"BS Management Engineering",
    }

#function for cleaniing year data
def clean_year(year):
    
    if (re.match(years,year)):
        year = "19" + year
        return year
    else:
        return year

#function for clesning course data
def clean_course(course): 
    for c in courses:
        if course==c:
            return str(courses[c])
            continue
        else:
            continue

#using pandas and lambda function for cleaning phone number data
recweek["phonenumber"]=recweek["phonenumber"].astype(str).apply(lambda x: '0'+x[-12:-9]+''+x[-9:-6]+''+x[-6:])

#vectorize
recweek["birthyear"] = recweek["birthyear"].apply(clean_year)  
recweek["course"] = recweek["course"].apply(clean_course)

#dump
recweek
```

```python
# write code here

#for both instances
#given all numbers are globe and smart
genders = ["Female","Male"]
globe_number = r"^09[1-2]7"
smart_number = r"0918"
def carrier(number):
    if (re.match(globe_number,number)):
        return "Globe"
    else:
        return "Smart"
cell_carrier = final_recweek[["gender","phonenumber"]].copy()
cell_carrier["carrier"] = cell_carrier["phonenumber"].apply(carrier)
carriers = cell_carrier["carrier"].unique()

#1st
gender_with_carrier = []

for c in carriers:
    for g in genders:
        gender_carrier_count = cell_carrier[cell_carrier["carrier"].str.contains(c) & cell_carrier["gender"].str.contains(g)]
        pair = [c,g,len(gender_carrier_count)]
        gender_with_carrier.append(pair)

gender_carrier_count = pd.DataFrame(gender_with_carrier,columns=["carrier","gender","count"])

#2nd
carrier = []

for c in carriers:
    count = cell_carrier[cell_carrier["carrier"].str.contains(c)]
    pair = [c,len(count)]
    carrier.append(pair)

carrier_count = pd.DataFrame(carrier,columns=["carrier","count"])

#dump
gender_carrier_count
```
